# Ejercicio 12

12. Agregar una columna a penguins con la mediana de la masa corporal de los ping¨uinos de cada
especie usando group_by() y agg().

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
median_species = penguins.groupby('species')['body_mass_g'].agg(np.median).reset_index()
median_species.rename(columns={'body_mass_g':'median_bodymass_g'}, inplace=True)
penguins.merge(median_species, on='species', how='left')
```

```python
#Otra opcion usando transform
penguins['median_bodymass_g'] = penguins.groupby('species')['body_mass_g'].transform(np.median)
```

13. Empezando con penguins, escribir una secuencia de operaciones que:

(a) Excluya a los ping¨uinos observados en la isla Biscoe.

(b) S´olo se quede con las variables que est´an entre species y body_mass_g inclusive.

(c) Renombre la variable species a especie_pinguino.

(d) Agrupe los datos por la variable especie_pinguino.

(e) Calcule el valor medio de las variables que contienen el string “length”, separando por
la especie del ping¨uino, y llamando a las columnas como las originales pero agregando
“ mean” al final.


```python
penguins = sns.load_dataset("penguins")
```

```python
#(a)
penguins=penguins.loc[[x for x in penguins.index if penguins['island'][x]!='Biscoe'],:]

#(b)
penguins = penguins.loc[:,'species':'body_mass_g']

#(c)
penguins= penguins.rename(columns={'species':'especie_pinguino'})

#(d)
penguins.groupby('especie_pinguino')
```

```python
#(e)
#Filtro las columnas con lenght
peng_length = [col for col in penguins.columns if 'length' in col]

#Agrupo por especie pinguino y calculo la media
mean_length = penguins.groupby('especie_pinguino')[peng_length].mean().reset_index()

#Renombro las columnas para incluir mean al final
mean_length.columns=[col + '_mean' if col in peng_length else col for col  in mean_length.columns]

print(mean_length)
```

```python
# Otra opcion para (b)
start= penguins.columns.get_loc('species')
end= penguins.columns.get_loc('body_mass_g')+1
penguins= penguins.iloc[:,start:end]
```

```python
peng_Adelie= penguins[penguins['species']=='Adelie']
peng_Adelie['body_mass_g'].mean()
peng_Adelie_Dream= peng_Adelie[peng_Adelie['island']=='Dream']
peng_Adelie_Dream['body_mass_g'].mean()
```
